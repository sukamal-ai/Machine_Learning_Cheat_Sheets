# Merge

```python
d1 = {"name":["Kalpesh","Kiran","Nimesh","Jitesh"],"math":[77,76,89,87]}
```

```python
d2 = {"name":["Kalpesh","Jitesh","Nimesh","Sushma"],"science":[79,86,82,84]}
```

```python
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2, on="name") #inner
```

```python
pd.merge(df1,df2, on="name", how="outer")
```

```python
pd.merge(df1,df2, on="name", how="right")
```

```python
pd.merge(df1,df2, on="name", how="left")
```

# Concat

```python
df_a = pd.DataFrame(np.arange(10).reshape(5,2),columns=["A","B"])
df_b = pd.DataFrame(np.arange(10,20).reshape(5,2),columns=["X","Y"])
```

```python
df_a
```

```python
df_b
```

```python
pd.concat([df_a,df_b],axis=1)
```

```python
pd.concat([df_a,df_b],axis=0)
```
