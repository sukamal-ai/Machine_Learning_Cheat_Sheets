# Numpy functions

```python
a = np.array([1,1,2,2,2,2,2,3,3,3,4,4,5,6,7,8,9])
```

```python
a
```

```python
np.mean(a)
```

```python
np.median(a)
```

```python
q1 = np.quantile(a,0.25)
```

```python
q3 = np.quantile(a,0.75)
```

```python
iqr = q3 - q1 
```

```python
iqr
```

```python
upper_range = q3 + 1.5*iqr
```

```python
lower_range = q1 - 1.5*iqr
```

```python
upper_range
```

```python
lower_range
```
