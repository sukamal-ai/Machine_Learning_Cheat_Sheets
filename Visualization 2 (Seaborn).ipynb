# Visualization 2 (Seaborn)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cars.csv")
```

## 1) Line plot

```python
x = np.arange(10)
y = np.sqrt(x)

plt.figure(figsize=(6,6))
sns.lineplot(x,y)
plt.show()
```

## 2) Histogram (Distplot)

```python
plt.figure(figsize=(6,6))
sns.distplot(df["price"])
plt.show()
```

```python
width_mean = df["width"].mean()
width_median = df["width"].median()

plt.figure(figsize=(6,6))
sns.distplot(df["width"])

plt.axvline(width_mean,color="red")
plt.axvline(width_median,color="yellow")
plt.show()
```

## 3) Count plot

```python
plt.figure(figsize=(6,6))
sns.countplot(data=df, x="body-style")
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(data=df, y="make")
plt.show()
```

```python
# Adding hue
plt.figure(figsize=(6,6))
sns.countplot(data=df, x="body-style",hue="fuel-type")
plt.show()
```

## 4) Heatmap

```python
corr = df.corr()
```

```python
plt.figure(figsize=(7,7))
sns.heatmap(corr,annot=True)
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.heatmap(corr,annot=True,cmap="viridis")
plt.show()
```

## 5) Scatter plot

```python
plt.figure(figsize=(7,7))
sns.scatterplot(data=df,x="width",y="price")
plt.plot()
```

```python
plt.figure(figsize=(7,7))
sns.scatterplot(data=df,x="width",y="price",hue="body-style")
plt.plot()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data=df,x="width",y="price",hue="make")
plt.plot()
```

## 6) Box plot

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=df, x="price")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=df, x="price",y="make")
plt.show()
```

## 7) Sub plots

```python
x = [1,2,3,4]
y = [3,4,5,6]

fig,axes = plt.subplots(2,2,figsize=(8,6))

axes[0,0].plot(x,y)
axes[0,1].scatter(x,y)
axes[1,0].plot(x,y)
axes[1,1].scatter(x,y)

plt.show()
```

```python
fig,axes = plt.subplots(2,2,figsize=(16,10))
sns.distplot(df["price"], ax=axes[0,0])
sns.countplot(data=df, y="make",ax=axes[0,1])
sns.scatterplot(data=df,x="width",y="price",hue="body-style",ax=axes[1,0])
sns.boxplot(data=df, x="price",ax=axes[1,1])
plt.show()
```
